```python
import os
import glob

import numpy as np
import pandas as pd
import geopandas as gpd

from sklearn.neighbors import KDTree
```

```python
aws_site_info = pd.read_csv("C:/Users/user/Desktop/DB/AWS/META_관측지점정보_20240930094934.csv", encoding = 'cp949')
```

```python
aws_site_info = aws_site_info.drop_duplicates(subset='지점')
```

```python
obs_files = glob.glob("./OUTPUT/*csv")
dongbu_d_df = pd.read_csv(obs_files[0])
```

```python
road_info = aws_site_info[['위도','경도','지점']]
```

```python
dongbu_d_df
```

```python
observ_line = np.array(list(zip(dongbu_d_df['lon'], dongbu_d_df['lat'])))
model_line = np.array(list(zip(aws_site_info['경도'], aws_site_info['위도'])))

tree = KDTree(observ_line)
# 각 점마다 가장 가까운 점을 찾아 매칭

matched_points = []
for point in model_line:
    _, index = tree.query([point], k=1)  # k=1로 설정하여 가장 가까운 점 하나만 선택
    matched_points.append(observ_line[index[0]])

matched_lon = [point[0][0] for point in matched_points]
matched_lat = [point[0][1] for point in matched_points]

df = pd.DataFrame({'경도': model_line[:, 0], '위도': model_line[:, 1],
                    'lon': matched_lon, 'lat': matched_lat})

```

```python
unique_coords = df[['lon', 'lat','경도','위도']].drop_duplicates()
```
