```python
import seaborn as sns
import matplotlib.pyplot as plt
dongbu_list = glob.glob("./OUTPUT/*csv")
for i in dongbu_list:

    file_name = i.split('\\')[-1].split(".")[0]
    # label    
    color_map = {
        "jr": "#636EFA",
        "mg": "#EF553B",
        "org": "#00CC96",
        "yc": "#AB63FA",
        "ss": "#FFA15A"
    }

    #   
    plt.figure(figsize=(16, 9))


    df = pd.read_csv(i)
    df = df[df['longitude'] > 1]
    #  
    sns.scatterplot(
        data=df, 
        x="longitude", 
        y="latitude", 
        hue="closest_column", 
        palette=color_map,  #  label   .
        edgecolor='none',  #   .
        s= 5
    )

    #   
    plt.title("Scatter plot with Surface Temperature")

    #  
    # plt.show()
    plt.savefig(f"./IMG/{file_name}.png",dpi = 200)
    plt.clf()
```
