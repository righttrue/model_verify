```python
import os
import glob

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import plotly.express as px
import plotly.graph_objects as go
```

```python
dongbu_list = glob.glob("*oly*.csv")
print(dongbu_list[4])
df = pd.read_csv(dongbu_list[4])
df = df[df['longitude'] > 1]
```

```python
# 첫 번째 산점도 생성
fig = px.scatter(df, x="longitude", y="latitude", color="surface_temperature",title="Scatter plot with Surface Temperature")#,range_color=(0, 10))


# 그래프 레이아웃 크기 조정
fig.update_layout(
    width=1600,  # 그래프 너비 (픽셀 단위)
    height=900  # 그래프 높이 (픽셀 단위)
)

# 그래프 표시
# 그래프 표시
fig.show()

# 첫 번째 산점도 생성
fig2 = px.scatter(df, x="longitude", y="latitude", color="closest_column",title="Scatter plot with Surface Temperature")#,range_color=(0, 10))


# 그래프 레이아웃 크기 조정
fig2.update_layout(
    width=1600,  # 그래프 너비 (픽셀 단위)
    height=900  # 그래프 높이 (픽셀 단위)
)

# 그래프 표시
fig2.show()
```

```python
import plotly.express as px

# label에 대한 고정 색상 매핑
color_map = {
    "jr": "#636EFA",
    "mg": "#EF553B",
    "org": "#00CC96",
    "yc": "#AB63FA",
    "ss": "#FFA15A"
}

# 산점도 생성
fig2 = px.scatter(
    df, 
    x="longitude", 
    y="latitude", 
    color="closest_column", 
    title="Scatter plot with Surface Temperature",
    color_discrete_map=color_map  # 각 label에 대해 색상을 고정합니다.
)

# 그래프 레이아웃 크기 조정
fig2.update_layout(
    width=1600,  # 그래프 너비 (픽셀 단위)
    height=900  # 그래프 높이 (픽셀 단위)
)

# 그래프 표시
fig2.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
dongbu_list = glob.glob("./OUTPUT/*csv")
for i in dongbu_list:

    file_name = i.split('\\')[-1].split(".")[0]
    # label에 대한 고정 색상 매핑
    color_map = {
        "jr": "#636EFA",
        "mg": "#EF553B",
        "org": "#00CC96",
        "yc": "#AB63FA",
        "ss": "#FFA15A"
    }

    # 그래프 크기 설정
    plt.figure(figsize=(16, 9))


    df = pd.read_csv(i)
    df = df[df['longitude'] > 1]
    # 산점도 생성
    sns.scatterplot(
        data=df, 
        x="longitude", 
        y="latitude", 
        hue="closest_column", 
        palette=color_map,  # 각 label에 대해 색상을 고정합니다.
        edgecolor='none',  # 점의 테두리를 제거합니다.
        s= 5
    )

    # 그래프 제목 설정
    plt.title("Scatter plot with Surface Temperature")

    # 그래프 표시
    # plt.show()
    plt.savefig(f"./IMG/{file_name}.png",dpi = 200)
    plt.clf()
```
