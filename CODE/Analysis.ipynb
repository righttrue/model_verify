## 오차가 가장 작은 점 매칭

```python
import seaborn as sns
import matplotlib.pyplot as plt
dongbu_list = glob.glob("./OUTPUT/*csv")
for i in dongbu_list:

    file_name = i.split('\\')[-1].split(".")[0]
    # label에 대한 고정 색상 매핑
    color_map = {
        "jr": "#636EFA",
        "mg": "#EF553B",
        "org": "#00CC96",
        "yc": "#AB63FA",
        "ss": "#FFA15A"
    }

    # 그래프 크기 설정
    plt.figure(figsize=(16, 9))


    df = pd.read_csv(i)
    df = df[df['longitude'] > 1]
    # 산점도 생성
    sns.scatterplot(
        data=df, 
        x="longitude", 
        y="latitude", 
        hue="closest_column", 
        palette=color_map,  # 각 label에 대해 색상을 고정합니다.
        edgecolor='none',  # 점의 테두리를 제거합니다.
        s= 5
    )

    # 그래프 제목 설정
    plt.title("Scatter plot with Surface Temperature")

    # 그래프 표시
    # plt.show()
    plt.savefig(f"./IMG/{file_name}.png",dpi = 200)
    plt.clf()   
```
