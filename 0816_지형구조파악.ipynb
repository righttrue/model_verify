```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Example DataFrame
# df = pd.DataFrame({'x': [...], 'y': [...], 'z': [...])

# Create a 3D plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the data
ax.plot(df['lon'], df['lat'], df['surface_temperature'], marker='o', linestyle='None')
# Plot the data
ax.plot(df['lon'], df['lat'], df['road_temp'], marker='x', linestyle='None')

# Set labels
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

# Display the plot
plt.show()
```

```python
import pandas as pd
import plotly.graph_objs as go
import plotly.io as pio

# Create the first trace
trace1 = go.Scatter3d(
    x=df['lon'],
    y=df['lat'],
    z=df['surface_temperature'],
    mode='markers',
    marker=dict(
        size=5,
        color='blue',
        opacity=0.8
    ),
    name='Surface Temperature'
)

# Create the second trace
trace2 = go.Scatter3d(
    x=df['lon'],
    y=df['lat'],
    z=df['road_temp'],
    mode='markers',
    marker=dict(
        size=5,
        color='red',
        opacity=0.8
    ),
    name='Road Temperature'
)

# Create the layout
layout = go.Layout(
    scene=dict(
        xaxis_title='Longitude / Locn',
        yaxis_title='Latitude',
        zaxis_title='Temperature'
    ),
    legend=dict(
        x=0,
        y=1
    )
    ,
    width=1200,  # Set the width of the figure
    height=1200   # Set the height of the figure
)

# Create the figure
fig = go.Figure(data=[trace1, trace2], layout=layout)

# Show the plot
pio.show(fig)

```
